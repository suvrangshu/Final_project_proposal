```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load the both test and train data into dataframe
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
#Display the size of both test and train data.
print("\nThe train data size is : {} ".format(train.shape)) 
print("The test data size is : {} ".format(test.shape))
```

```python
train.head()
```

```python
#Concatinate both dataframe to find how many missing cells in each columns
all_data = pd.concat((train, test)).reset_index(drop=True)

#Get the sum of all the empty values in the dataframe
all_data_na = all_data.isnull().sum()  

#Ignore the columsn which don't have any empty cells 
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)[:30]

#Assign it to a data frame and display the count
missing_data = pd.DataFrame({'Missing Ratio' :all_data_na})
missing_data
```

```python
#Getting the count of both categorical and numerical columns
cat = len(train.select_dtypes(include=['object']).columns)
num = len(train.select_dtypes(include=['int64','float64']).columns)
print('Total Features: ', cat, 'categorical', '+',
      num, 'numerical', '=', cat+num, 'features')
```

```python
#Remove the ID column
train.drop(['Id'], axis=1)

#Replace the null cells in the numerical column with 0
train_numerical = train.select_dtypes(exclude=['object'])
train_numerical.fillna(0,inplace = True)

#Replace the null cells in the catogorical column with NONE
train_categoric = train.select_dtypes(include=['object'])
train_categoric.fillna('NONE',inplace = True)

#Merge both the column. Now the Train dataframe don't have any empty cells
train = train_numerical.merge(train_categoric, left_index = True, right_index = True)
```
